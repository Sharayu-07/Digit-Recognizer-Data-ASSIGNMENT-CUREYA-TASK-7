```python
mnist = tf.keras.datasets.mnist #handwritten characters based on 28*28 sized images of 0 to 9
```

```python
(X_train, y_train),(X_test,y_test) = mnist.load_data()
#Dividing and unpacking the dataset into train and test datasets
```

# Pre - Processing ( Normalization ) :

```python
X_train = tf.keras.utils.normalize (X_train, axis = 1) # Normalizing the Data as images are in Gray color not colored (RGB) all values varies from 0 to 255
X_test = tf.keras.utils.normalize (X_test, axis = 1)
plt.imshow(X_train[0], cmap = plt.cm.binary)
```

# After Normalizing :

```python
print (X_train[0]) #After Normalizing (X_train = Data)
```

```python
print (y_train[0]) #After Normalizing (y_train = Label)
```

# Resizing the Image :

```python
import numpy as np 
IMG_SIZE = 28
X_trainr = np.array(X_train).reshape(-1, IMG_SIZE, IMG_SIZE, 1) # increasing 1 dimension for kernel = filter operation
X_testr = np.array(X_test).reshape(-1, IMG_SIZE, IMG_SIZE, 1) # increasing 1 dimension for kernel operation
print ("Training Samples Dimension", X_trainr.shape)
print ("Testing Samples Dimension", X_testr.shape)
# To make it suitable for apply convolution operation, resizing the image is must
```

```python
from keras.models import Sequential
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
from keras.layers import Activation, Dense, Flatten
```

```python
model = Sequential()
# Creating a Neural Network

model.add(Conv2D(64, (3, 3), input_shape = X_trainr.shape[1:])) # First Convolution Layer
model.add(Activation("relu")) # Making it Non - Linear
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, (3, 3))) # Second Convolution Layer
model.add(Activation("relu")) 
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, (3, 3))) # Third Convolution Layer
model.add(Activation("relu")) 
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Flatten()) # Fully Connected Layer 1, need to be flattened to be 2D to 1D
model.add(Dense(64))
model.add(Activation("relu"))

model.add(Dense(32)) # Fully Connected Layer 2
model.add(Activation("relu"))

model.add(Dense(10)) # Last Fully Connected Layer, output must be equal to number of classes 10 (0-9) 
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
print("Total Training Samples = ",len(X_trainr))
```

```python
model.compile(loss = "sparse_categorical_crossentropy", optimizer = "adam", metrics = ['accuracy'])
```

# Model Training 

```python
model.fit(X_trainr, y_train, epochs = 5, validation_split = 0.3)
```

```python
test_loss, test_acc = model.evaluate(X_testr, y_test)
print ("Test loss on 10,000 test samples", test_loss)
print ("Validation accuracy on 10,000 test samples", test_acc)
```
